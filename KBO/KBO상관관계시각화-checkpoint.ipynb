```python
def split_dataframe(df):
    return [df.iloc[:, i:i+2] for i in range(0, df.shape[1], 2)]

# DataFrame을 두 열씩 분할
split_dfs = split_dataframe(df)

# 찾고자 하는 문자열 목록
search_terms = ['홀드', '세이브', '퀄리티스타트', '풋아웃', '평균연봉', '누적이동거리']

# 분할된 DataFrame을 순회하며 조건에 맞는 값 추출
for i, split_df in enumerate(split_dfs):
    print(f"DataFrame {i + 1}:\n{split_df}\n")
    
    # 조건에 맞는 행 필터링
    matched_rows = split_df[split_df.iloc[:, 0].isin(search_terms)]
    
    # 두 번째 열의 값 추출
    if not matched_rows.empty:
        matched_values = matched_rows.iloc[:, 1].tolist()
        print(f"Matched values from DataFrame {i + 1}: {matched_values}\n")
    else:
        print(f"No matches found in DataFrame {i + 1}.\n")
```
